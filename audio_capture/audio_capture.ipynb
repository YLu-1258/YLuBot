```python
import pyaudio
import numpy as np
import wave
import keyboard

```

```python
def audio_capture():
    recorded_snippets = []
    while True:
        if keyboard.is_pressed('space'):
            break
        data = record_audio()
        total_sum = sum(np.sum(np.abs(arr)) for arr in data)
        if (total_sum > 200000):
            recorded_snippets.append(data)
        else:
            pass
    return recorded_snippets

def record_audio():
    chunk = 1024
    sample_format = pyaudio.paInt16
    channels = 1
    fs = 44100
    seconds = 3

    recorder = pyaudio.PyAudio()  # Create an interface to PortAudio

    print('Recording')

    stream = recorder.open(format=sample_format,
                    channels=channels,
                    rate=fs,
                    frames_per_buffer=chunk,
                    input=True)

    frames = []  # Initialize array to store frames


    counter = 0
    while (counter < 25):
        data = stream.read(chunk)
        numpydata = np.frombuffer(data, dtype=np.int16)
        frames.append(numpydata)
        if np.sum(np.abs(numpydata)) < 4000:
            counter+=1
        else:
            counter = 0
            
    return frames
```

```python
frames = audio_capture()
frames
```

```python
# Save the recorded data as a WAV file
for i in range(len(frames)):
    wf = wave.open(f'test_audio/{i+1}.wav', 'wb')
    wf.setnchannels(1)
    wf.setsampwidth(pyaudio.PyAudio().get_sample_size(pyaudio.paInt16))
    wf.setframerate(44100)
    wf.writeframes(b''.join(frames[i]))
    wf.close()
```
